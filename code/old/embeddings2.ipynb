```python
import json
import pandas as pd

# Grabbing proxifier logs

proxifier_log_fie = "/content/code-and-datasets/data/summary/logs/Proxifier.txt"

SUMMARY = 'summary'
LOG = 'log'
def get_record_type(key):
  if key.startswith("#summary:#"):
    return SUMMARY
  else:
    return LOG

def load_data(file):
  data_log = {}
  data_summary = []
  current_key = None
  lines = []
  record_type = LOG

  with open(file, 'r') as file:
    for line in file:
      line = line.strip()
      if not line:
        continue

      if line.startswith("#"):
        if current_key is not None:
          if get_record_type(current_key) == SUMMARY:
            data_summary.append(lines)
          else:
            data_log[current_key.replace("#", "")] = lines

        current_key = line
        lines = []
      else:
        lines.append(line)

    if current_key is not None:
      data_summary.append(lines)
  return data_log, data_summary

p_data, p_summary = load_data(proxifier_log_fie)
#print(p_data['1'])

a = ""
for i in range(0, len(p_data['1'])):
  a = a + p_data['1'][i] + '\n '
print(a)

print(p_summary[0][0])

df = pd.DataFrame(columns=[ 'log_data', 'log_summary', 'product'])
for i in range(1, len(p_data)+1):
  #print(p_data[str(i)])
  #print(p_summary[i-1][0])
  string_data = ""
  for j in range(0, len(p_data[str(i)])):
    string_data = string_data + p_data[str(i)][j] + "\n "
  df = df.append({'log_data': string_data, 'log_summary':p_summary[i-1][0], 'product':"Proxifier"} , ignore_index=True)

print(df.items)
```

```python
def open_ai_zk1_summary(vectorstore, model, input):
  #print(input)
  input = '\n'.join(input)
  prompts=[]
  prompts.append({"role": "system", "content": "You are a zookeeper specialist"})
  prompts.append({"role": "user", "content": "Help me to create a concise summary less than 20 words, \
  the output is a json summary with the following keys: summary, anomaly"})
  prompts.append({"role": "assistant", "content": "yes, I can do it."})
  results = vectorstore.similarity_search_with_score(input, k=5)
  time.sleep(1)



  for i, j in enumerate(results):
    userContent = j[0].page_content

    assistantContent = j[0].metadata['log_summary']
    #print(assistantContent)

    prompts.append({"role": "user", "content": f"""create a summary for: {userContent}"""})
    #data = f""""summary": {assistantContent}"""
    prompts.append({"role": "assistant", "content": assistantContent})

  # print(input)
  prompts.append({"role": "user", "content": f"create a summary for: {input}"})
  #print(prompts)


  response = client.chat.completions.create(
    model=model,
    messages=prompts,
    temperature=0.3,
    max_tokens=250,
    top_p=1,
    frequency_penalty=0,
    presence_penalty=0,
    stop=None
  )
  #print(response)

  #json_object = json.loads(response.choices[0].message.content)
  #print(json_object['summary'])
  print(response.choices[0].message.content.replace('\n', ''))
  return response.choices[0].message.content.replace('\n', '')
  #return json_object['summary']


```
